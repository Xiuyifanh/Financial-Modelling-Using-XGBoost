```python
# Fetch AAPL stock data
ticker = "AAPL"
data = yf.download(ticker, start="2022-05-01", end="2024-11-25")  # Adjust dates if needed

# Feature Engineering
data['Return'] = data['Close'].pct_change()
data['SMA_10'] = data['Close'].rolling(window=10).mean()
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_100'] = data['Close'].rolling(window=100).mean()
data['Volatility'] = data['Return'].rolling(window=16).std()
data['EMA_10'] = data['Close'].ewm(span=10, adjust=False).mean()
data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()
data['RSI'] = 100 - (100 / (1 + data['Return'].rolling(window=10).mean() / data['Return'].rolling(window=14).std()))
data.dropna(inplace=True)

# Define target variable
data['Target'] = np.log1p(data['Close'].shift(-1))  # Apply log1p for better scaling
data.dropna(inplace=True)
```

```python
# Split data into training and testing sets
train_data = data[0:int(len(data) * 0.84)]
test_data = data[int(len(data) * 0.84):]

# Define features and targets
X_train = train_data[['Close', 'SMA_10', 'SMA_50', 'SMA_100', 'Volatility', 'EMA_10', 'EMA_50', 'RSI']].values
y_train = train_data['Target'].values
X_test = test_data[['Close', 'SMA_10', 'SMA_50', 'SMA_100', 'Volatility', 'EMA_10', 'EMA_50', 'RSI']].values
y_test = test_data['Target'].values
```

```python
# Scale features and target variables
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).flatten()
y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).flatten()
```

```python
# Define initial XGBoost model
xgboost_model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=1000, learning_rate=0.05, max_depth=4)

# Define parameter grid for GridSearch
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 500],
    'max_depth': [3, 4, 5]
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=xgboost_model, param_grid=param_grid, cv=3, n_jobs=-1)
grid_search.fit(X_train_scaled, y_train_scaled)

# Retrieve the best parameters
best_params = grid_search.best_params_
print(f"Best Parameters: {best_params}")

# Train model with best parameters
xgb_model_best = xgb.XGBRegressor(objective='reg:squarederror',
                                  learning_rate=best_params['learning_rate'],
                                  n_estimators=best_params['n_estimators'],
                                  max_depth=best_params['max_depth'])
xgb_model_best.fit(X_train_scaled, y_train_scaled)
```

```python
# Perform feature selection
sfm = SelectFromModel(xgb_model_best, threshold=0.01)
sfm.fit(X_train_scaled, y_train_scaled)

# Apply selected features
X_train_selected = sfm.transform(X_train_scaled)
X_test_selected = sfm.transform(X_test_scaled)

# Retrain model with selected features
xgb_model_best.fit(X_train_selected, y_train_scaled)
```

```python
# Predictions
y_pred_train_scaled = xgb_model_best.predict(X_train_selected)
y_pred_test_scaled = xgb_model_best.predict(X_test_selected)

# Inverse scale and apply inverse log1p
y_pred_train = np.expm1(scaler_y.inverse_transform(y_pred_train_scaled.reshape(-1, 1)).flatten())
y_pred_test = np.expm1(scaler_y.inverse_transform(y_pred_test_scaled.reshape(-1, 1)).flatten())

# Actual prices
y_train_actual = np.expm1(y_train)
y_test_actual = np.expm1(y_test)

# Evaluate model
r2_test = r2_score(y_test_actual, y_pred_test)
mse_test = mean_squared_error(y_test_actual, y_pred_test)
mae_test = mean_absolute_error(y_test_actual, y_pred_test)
mape_test = mean_absolute_percentage_error(y_test_actual, y_pred_test)

print(f"R2 on Test Data: {r2_test}")
print(f"MSE on Test Data: {mse_test}")
print(f"MAE on Test Data: {mae_test}")
print(f"MAPE on Test Data: {mape_test * 100:.2f}%")
```

```python
# Plot Actual vs Predicted
plt.figure(figsize=(12, 6))
plt.plot(test_data.index, y_test_actual, label="Actual Prices", color='blue')
plt.plot(test_data.index, y_pred_test, label="Predicted Prices", color='red')
plt.xlabel("Date")
plt.ylabel("AAPL Stock Price")
plt.title("AAPL Stock Price Prediction")
plt.legend()
plt.show()

# Scatter Plot: Actual vs Predicted
plt.figure(figsize=(8, 8))
plt.scatter(y_test_actual, y_pred_test, color='red', alpha=0.6, label='Predicted vs Actual')
plt.plot([min(y_test_actual), max(y_test_actual)], [min(y_test_actual), max(y_test_actual)], color='black', linestyle='dotted')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
plt.legend()
plt.grid()
plt.show()
```

```python
# Forecast the next 7 trading days
last_data = test_data[['Close', 'SMA_10', 'SMA_50', 'SMA_100', 'Volatility', 'EMA_10', 'EMA_50', 'RSI']].iloc[-1].values
forecast_prices = []

for _ in range(7):
    # Scale and select features
    last_data_scaled = scaler_X.transform(last_data.reshape(1, -1))
    last_data_selected = sfm.transform(last_data_scaled)

    # Predict the next price
    forecast_price_scaled = xgb_model_best.predict(last_data_selected)
    forecast_price = np.expm1(scaler_y.inverse_transform(forecast_price_scaled.reshape(-1, 1)).flatten()[0])
    forecast_prices.append(forecast_price)

    # Update last_data for the next prediction
    last_data = np.roll(last_data, -1)
    last_data[-1] = forecast_price

# Generate forecast dates
forecast_dates = pd.date_range(test_data.index[-1], periods=8, freq='B')[1:]

# Display forecast
print("Forecast for the next 7 trading days:")
for date, price in zip(forecast_dates, forecast_prices):
    print(f"{date.strftime('%Y-%m-%d')}: ${price:.2f}")
```
